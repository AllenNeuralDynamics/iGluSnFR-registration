```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

# Extract the motionDSc data
motionDSc_mat = loadmat('/root/capsule/scratch/Motion-michael.mat')['motionDSc'].squeeze()

# Extract the motionDSc data
motionDSr_mat = loadmat('/root/capsule/scratch/Motion-michael.mat')['motionDSr'].squeeze()

# Path to the HDF5 file
alignmentData_h5_path = '/root/capsule/results/1/scan_00002_20240720_161305_ALIGNMENTDATA.h5'

# Open the HDF5 file in read mode
with h5py.File(alignmentData_h5_path, "r") as f:
    # Extract motionC and motionR datasets
    motionDSc = f['aData/motionDSc'][:]
    motionDSr = f['aData/motionDSr'][:]

    print('RMSE motionDSc--->', np.sqrt(np.mean((motionDSc - motionDSc_mat) ** 2)))
    print('RMSE motionDSr--->', np.sqrt(np.mean((motionDSr - motionDSr_mat) ** 2)))

    # Set the figure size (width, height) in inches
    plt.figure(figsize=(50, 12))

    # Plot the data with specified colors and line styles
    plt.plot(motionDSc[-1500:], color='blue', label='motionDSc', linestyle='dashdot', linewidth=0.9)
    plt.plot(motionDSc_mat[-1500:], color='red', linestyle='--', label='motionDSc_mat', linewidth=0.9)

    # Add gridlines
    plt.grid(True)

    # Add a legend
    plt.legend()

    # Show the plot
    plt.show()
```

# NormCorr

```python
from jnormcorre.motion_correction import MotionCorrect
from scipy.io import loadmat
import tifffile
import matplotlib.pyplot as plt
```

```python
Yhp = loadmat('/root/capsule/scratch/For_NormCorre.mat')['Yhp']
best_cluster = loadmat('/root/capsule/scratch/For_NormCorre.mat')['best_cluster']
maxshift = 50
initFrames = 1000
```

```python
# Initialize the MotionCorrect object with all necessary parameters
corrector = MotionCorrect(
    lazy_dataset = Yhp.transpose(2, 0, 1),              # Transpose to match (frame, height, width)
    max_shifts=(maxshift, maxshift),      # Maximum allowed shifts in pixels
    pw_rigid=False,                       # Piecewise rigid correction disabled
    frames_per_split=initFrames,           # Ensures consistency with MATLAB's 'bin_width'
    upsample_factor_grid = 4,
    niter_rig=1,  # Number of iterations for rigid correction
)

frame_corrector, output_file = corrector.motion_correct(
    template=np.mean(Yhp[:, :, list(best_cluster.ravel()- 1)], axis=2).squeeze(), 
    save_movie=True
)
```
